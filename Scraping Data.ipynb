```python
from bs4 import BeautifulSoup
import re
import datetime


date = "1-16-17"


input_html = "1-16-17.html"
with open(input_html, "r") as ifile:
    soup = BeautifulSoup(ifile, 'html5lib')
```

```python
def check_date(year, month, day):
    correctDate = None
    try:
        newDate = datetime.datetime(year, month, day)
        correctDate = True
    except ValueError:
        correctDate = False
    return correctDate
```

```python
from bs4 import BeautifulSoup
import re
import datetime
import requests



first_half = "https://www.wunderground.com/history/airport/KNYC/2017/"

second_half = "/DailyHistory.html"

write_file = ""


for month in range(1, 13):
    for day in range (1, 32):
        if (check_date(2017, month, day)):
            date = "2017" + "-" + str(month) + "-" + str(day)
            print(date)
            with open(input_html, "r") as ifile:
                website = first_half + str(month) + "/" + str(day) + second_half
                url = requests.get(website)
                page = url.text

                soup = BeautifulSoup(page, 'html5lib')
            temp = soup.findAll("tr", {"class": "no-metars"})
            selected_data = ''
            for t in temp:
                selected_data += t.get_text() +'AAAAAAAAAA\n\n\n' 
            selected_data = selected_data.replace("\n\t\t", ",").replace("\n", "").replace("&nnbsp", "").replace(",AAAAAAAAAA",'\n')
            tokens = selected_data.split("\n")
            for t in tokens:
                if (t != ""):
                    write_file += date + t + "\n"
text_file = open("weather.txt", "w")
text_file.write(write_file)
text_file.close()


```
